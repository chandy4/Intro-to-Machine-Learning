```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 

df = pd.read_csv(r'C:\Users\Connor\Documents\IntroToMLHW1\D3.csv', header=None)
df.head()
m=len(df)

x1 = df.values[:, 0]
x2 = df.values[:, 1]
x3 = df.values[:, 2]
y = df.values[:, 3]

x0 = np.ones((m, 1))
x1 = x1.reshape(m, 1)
x2 = x2.reshape(m, 1)
x3 = x3.reshape(m, 1)
```

```python
def computeCost(x, y, theta):
    predictions = x.dot(theta)
    errors = np.subtract(predictions, y)
    sqrErrors = np.square(errors)
    J = 1/(2 * m) * np.sum(sqrErrors)
    
    return J
```

```python
def gradientDescent(x, y, theta, alpha, iterations):
    cost_history = np.zeros(iterations)

    for i in range(iterations):
        predictions = x.dot(theta)
        errors = np.subtract(predictions, y)
        sum_delta = (alpha / m) * x.transpose().dot(errors)
        theta = theta - sum_delta;
        cost_history[i] = computeCost(x, y, theta)
        
    return theta, cost_history
    
```

```python
#Training for x1 

iterations = 1500
alpha = 0.1
x = np.hstack((x0, x1))
theta = np.zeros(2)

theta, cost_history = gradientDescent(x, y, theta, alpha, iterations)

print('Final value of theta =',theta)
print('cost_history =',cost_history)
```

```python
plt.scatter(x1, y, color = 'red', marker = "+", label = 'Training Data')
plt.plot(x1, x.dot(theta), color = 'green', label = 'Linear Regression')
plt.title('Linear Regression for x1')
plt.grid()
plt.legend()
```

```python
#Convergence of gradient descent for x1

iteration_array = np.arange(1, iterations + 1)

plt.plot(iteration_array, cost_history, color= 'purple', label = 'Cost')
plt.title('Covergence of Gradient Descent for x1')
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.grid()
```
